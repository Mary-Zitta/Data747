### Dataset
The dataset to be used for this lab assignment is the Titanic dataset, which contains information about passengers on the Titanic, including their demographics, ticket information, and survival status. The dataset can be downloaded from Kaggle: https://www.kaggle.com/c/titanic/data

You can load the dataset using the following code:

```python

import seaborn as sns

# Load the Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')
titanic.count()
```

Summary statistics

Distribution of target variable

```python
import matplotlib.pyplot as plt
target_counts = titanic['survived'].value_counts()

plt.bar(target_counts.index, target_counts.values, color=['red', 'green'])
plt.xlabel('Survival Status (0 = Died, 1 = Survived)')
plt.ylabel('Count')
plt.title('Distribution of Survival')
plt.show()

```

Distribution of survival by class using stacked bar chart

```python
sns.histplot(data=titanic, x='survived', hue='survived', multiple='stack')
plt.title('Stacked chart of Survival by Passenger Class')
plt.show()
```

Visual Histogram of Age

```python
sns.histplot(data=titanic, x='age')
plt.title('Age Distribution of Passengers')
plt.show()
```

Relationship between age and fare

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title('Relationship between age and fare')
plt.show()
```

Detailed report of Dataset

```python
from ydata_profiling import ProfileReport
profile = ProfileReport(titanic, title='Pandas Profiling Report', explorative=True)
profile.to_notebook_iframe()
```

Identify and visualize missing values

```python
sns.heatmap(titanic.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

titanic.isnull().sum()
```

Identify and visualize outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_variables = titanic.select_dtypes(include='number').columns

sns.boxplot(data=titanic[numeric_variables], orient='h', palette='Set2')
plt.title('Outlier Detection using boxplots')
plt.show()
```

Correlations between variables

```python

sns.pairplot(data=titanic)
plt.title('Outlier Detection using boxplots')
plt.show()
```
