5. Create a pie chart of the provided dataset with the percentage of car sales of each brand.

```python
import matplotlib.pyplot as plt
import pandas as pd

car_sales = pd.read_csv('C:\\Users\\Kihumba\\OneDrive\\Data747\\Data747\\Week 5\\car_sales.csv')

car_sales['Total'] = car_sales[['Petrol', 'Diesel', 'Hybrid', 'Electric']].sum(axis=1)

plt.figure(figsize=(6,6))
plt.pie(car_sales['Total'], labels=car_sales['Brand'], autopct='%1.1f%%', startangle=90)
plt.title('Percentage Car Sales by Brand')
plt.show()


```

6. Create a stacked bar chart of the provided dataset with the years on the x-axis and the total number of medals won by each country on the y-axis. The stacked bars should show the number of gold, silver, and bronze medals won.

```python
import matplotlib.pyplot as plt
import pandas as pd

medals = pd.read_csv('C:\\Users\\Kihumba\\OneDrive\\Data747\\Data747\\Week 5\\medal_counts.csv')

plt.figure(figsize=(9,6))
countries = medals['Country']
gold = medals['Gold']
silver = medals['Silver']
bronze = medals['Bronze']

plt.bar(countries, gold, label='Gold', color='green')
plt.bar(countries, silver, bottom=gold, label='Silver', color='blue')
plt.bar(countries, bronze, bottom=gold+silver, label='Bronze', color='orange')

for i, country in enumerate(medals['Country']):
    plt.text(i, medals.loc[i, 'Gold']/2, str(medals.loc[i, 'Gold']), ha='center', va='center', color='black', fontsize=10)
    plt.text(i, medals.loc[i, 'Gold'] + medals.loc[i, 'Silver']/2, str(medals.loc[i, 'Silver']), ha='center', va='center', color='black', fontsize=10)
    plt.text(i, medals.loc[i, 'Gold'] + medals.loc[i, 'Silver'] + medals.loc[i, 'Bronze']/2, str(medals.loc[i, 'Bronze']), ha='center', va='center', color='black', fontsize=10)

plt.title('Total Medals Won by Each Country')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.legend()
plt.show()
```

10. Create a 3D scatter plot of the provided dataset with horsepower on the x-axis, weight on the y-axis, and miles per gallon on the z-axis.
