Create a stacked area plot of the provided dataset with the years on the x-axis and the total sales of each car brand on the y-axis. The stacked areas should show the sales of cars with different fuel types (petrol, diesel, hybrid, electric).

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('p:\\Python Programming and Data Analysis\\data\\car_sales.csv')
df.set_index('Year', inplace=True)

fig, ax = plt.subplots()
ax.stackplot(df.index, df['Petrol'], df['Diesel'], df['Hybrid'], df['Electric'], labels=['Petrol', 'Diesel', 'Hybrid', 'Electric'])
ax.legend(loc='upper left')
plt.show()
```

Create a scatter plot of the provided dataset with horsepower on the x-axis and weight on the y-axis, with the marker size indicating the car price.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('p:\\Python Programming and Data Analysis\\data\\car_data.csv')

plt.scatter(df['Horsepower'], df['Weight'], s=df['Price'], alpha=0.5)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.show()
```

Create a horizontal bar chart of the provided dataset with the car brands on the y-axis and the total sales on the x-axis.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('p:\\Python Programming and Data Analysis\\data\\car_sales.csv')
df.set_index('Brand', inplace=True)

df['Total Sales'] = df.sum(axis=1)

df['Total Sales'].sort_values().plot(kind='barh')
plt.xlabel('Total Sales')
plt.ylabel('Car Brands')
plt.show()
```

Create a pie chart of the provided dataset with the percentage of medals won by each country.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('p:\\Python Programming and Data Analysis\\data\\medal_counts.csv')
df.set_index('Country', inplace=True)

df.plot(kind='pie', y='Total Medals', legend=None)
plt.ylabel('')
plt.show()
```

Create a heatmap of the provided dataset with the car brands on the y-axis and the average price, horsepower, and mileage on the x-axis.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('p:\\Python Programming and Data Analysis\\data\\car_data.csv')
df.set_index('Brand', inplace=True)

fig, ax = plt.subplots()
im = ax.imshow(df[['Price', 'Horsepower', 'Mileage']].T)

ax.set_xticks(range(len(df)))
ax.set_xticklabels(df.index)
ax.set_yticks(range(3))
ax.set_yticklabels(['Price', 'Horsepower', 'Mileage'])

plt.colorbar(im)
plt.show()
```
